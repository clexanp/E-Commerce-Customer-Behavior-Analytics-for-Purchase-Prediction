## Proses Analisis Missing Value pada Kolom Rate

```python
data_pelanggan[
    data_pelanggan["Rate"].isnull()
]["Action"].value_counts()
```

## Insight Proses Analisis Missing Value pada Kolom Rate

1. Dapat dilihat bahwa pola missing value pada kolom Rate sama seperti kolom Quantity yang sudah di analisis sebelumnya.
2. Pada aktivitas non-transaksi tidak memiliki informasi harga produk. Hal tersebut menunjukkan bahwa case dari missing value ini berasal dari proses bisnis yang normal.

## Proses Analisis Missing Value pada Kolom Total Price

```python
data_pelanggan[
    data_pelanggan["Total Price"].isnull()
]["Action"].value_counts()
```

## Insight Proses Analisis Missing Value pada Kolom Total Price

1. Kurang lebih hasil analisis nya sama dengan yang terjadi di kolom Quantity dan Rate, yaitu missing value hanya muncul pada aktivitas pelanggan yang belum menghasilkan transaksi.
2. Sehingga bisa dilanjutkan ke tahap berikutnya yaitu proses menangani missing value.

## Proses Menangani Missing Value

```python
data_pelanggan["Quantity"] = (
    data_pelanggan["Quantity"]
    .fillna(0)
)

data_pelanggan["Rate"] = (
    data_pelanggan["Rate"]
    .fillna(0)
)

data_pelanggan["Total Price"] = (
    data_pelanggan["Total Price"]
    .fillna(0)
)

data_pelanggan.isnull().sum()
```

## Insight Proses Menangani Missing Value

1. Missing value pada Quantity, Rate, dan Total Price diubah menjadi 0.
2. Nilai 0 sengaja dipilih karena aktivitas pada 3 kolom tersebut belum menghasilkan transaksi.
